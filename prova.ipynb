```python
import numpy as np
import matplotlib.pyplot as plt


r = 4
b = 32

s = np.linspace(0, 1, 500)

f_s = 1 - (1 - s**r)**b

s_point = 0.4
f_point = 1 - (1 - s_point**r)**b

plt.figure(figsize=(8, 5))
plt.plot(s, f_s, color = 'blue')
plt.scatter(s_point, f_point, color='red', zorder=5, label=f'Minimum similarity threshold ($s^*$)')
plt.xlabel('similarity')
plt.ylabel('p(s)')
plt.title(f'Plot of $p(s) = 1-(1-s^r)^b$ with r={r} and b={b}')
plt.grid(True)
plt.legend()
plt.savefig('curva.eps', format='eps')
plt.show()

```
